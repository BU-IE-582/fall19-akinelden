```python
import matplotlib.pyplot as pyplot
pyplot.imshow(pix)
pyplot.show()
```

```python
R=np.copy(pix)
R[:,:,1:]=0
```

```python
G=np.copy(pix)
G[:,:,[0,2]]=0
```

```python
B=np.copy(pix)
B[:,:,[0,1]]=0
```

```python
pyplot.imshow(R)
pyplot.show()
```

```python
pyplot.subplot(1,5,1)
pyplot.imshow(R)
pyplot.subplot(1,5,3)
pyplot.imshow(G)
pyplot.subplot(1,5,5)
pyplot.imshow(B)

pyplot.show()
```

```python
R_vec= np.array([])
G_vec= np.array([])
B_vec= np.array([])
```

```python
for i in range(0,512):
    R_avg = np.mean(R[:,i,0])
    R_vec=np.append (R_vec, R_avg)
```

```python
for i in range(0,512):
    G_avg = np.mean(G[:,i,1])
    G_vec=np.append (G_vec, G_avg)
```

```python
for i in range(0,512):
    B_avg = np.mean(B[:,i,2])
    B_vec=np.append (B_vec, B_avg)
```

```python
pyplot.plot(range(0,512), R_vec, color="red")
pyplot.plot(range(0,512), G_vec, color="green")
pyplot.plot(range(0,512), B_vec, color="blue")
pyplot.show()
```
